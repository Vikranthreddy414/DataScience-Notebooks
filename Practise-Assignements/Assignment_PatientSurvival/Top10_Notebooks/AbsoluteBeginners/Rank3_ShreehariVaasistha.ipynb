# **Data Processing**

```python
phdata1 = pd.get_dummies(pharma_data,columns=['Treated_with_drugs','Patient_Smoker','Patient_Rural_Urban', 'Patient_mental_condition'])
```

```python
phdata1.drop(['Patient_ID'], axis = 1, inplace = True)
```

# **DPhi Test Data**

```python
test_new = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Testing_set_begs.csv')
```

```python
test_new1 = pd.get_dummies(test_new,columns=['Treated_with_drugs','Patient_Smoker','Patient_Rural_Urban', 'Patient_mental_condition'])
```

```python
test_new1.drop(['Patient_ID'], axis=1, inplace=True)
```

# **Convert to train and test data**

```python
# Split data into Input and Target Variable
b = phdata1[['Survived_1_year']]  # target variable 
A = phdata1.drop(['Survived_1_year', 'Patient_Smoker_Cannot say'], axis=1)  # input variables
```

```python
# Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(A, b, test_size=0.33, random_state=1)
```

# **1. XGBoost**

**Note: XGBoost handles missing values better than i did. Performance was also great comparatively. Hence using it. Still i'm trying to handle missing values in a similar way.**

```python
from xgboost import XGBClassifier
from xgboost import plot_importance

# fit model to training data
xgb_model = XGBClassifier(booster =  'gbtree', eta = 0.8)
xgb_model.fit(X_train, y_train)
```

```python
predxgb = xgb_model.predict(test_new1)
```

```python
print(*predxgb, sep = ",")  
```

```python
from xgboost import XGBClassifier
from xgboost import plot_importance

# fit model to training data
x_model = XGBClassifier(booster =  'gbtree',
                        eta = 1,
                        max_depth = 5,
                        min_child_weight = 8, 
                        scale_pos_weight = 1.2
                        )
x_model.fit(X_train, y_train)

# plot feature importance
plot_importance(x_model)
plt.show()
```

```python
xp = x_model.predict(test_new1)
# print(*xp, sep=',')
```
