```
import numpy as np
import pandas as pd
```

```
data=pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Training_set_begs.csv')
data
```

```
data.Patient_mental_condition.value_counts()
```

```
data=data.drop('Patient_mental_condition',axis=1)
```

```
data.Z.value_counts()
```

```
data=data.drop('Z',axis=1)
```

```
data.info()
```

```
from scipy import stats
cols=['Treated_with_drugs','A','B','C','D','E','F']
for i in cols:
      data[i]=data[i].fillna(stats.mode(data[i])[0][0])
```

```
data.info()
```

```
for i in range(23097):  
    data.Number_of_prev_cond[i]=data.A[i]+data.B[i]+data.C[i]+data.D[i]+data.E[i]+data.F[i]
```

```
data.info()
```

```
for i in range(23097):
  if data.Patient_Smoker[i]=='Cannot say':
    data.Patient_Smoker[i]='NO'
```

```
data
```

```
data=pd.get_dummies(data,columns=['Treated_with_drugs','Patient_Smoker','Patient_Rural_Urban'])
```

```
data
```

```
X=data.drop('Survived_1_year',axis=1)
y=data[['Survived_1_year']]
```

```
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=9,test_size=0.2)
```

```
test_new=pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/pharma_data/Testing_set_begs.csv')
test_new=test_new.drop(['Patient_mental_condition','Z'],axis=1)
test_new=pd.get_dummies(test_new,columns=['Treated_with_drugs','Patient_Smoker','Patient_Rural_Urban'])
```

```
test_new
```

```
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score,f1_score
model=XGBClassifier(random_state=9,max_depth=4,gamma=0.7,subsample=1,colsample_bytree =1,scale_pos_weight =1,learning_rate=0.2,n_estimators=100).fit(X_train,y_train)
y_pred=model.predict(X_test)
print(accuracy_score(y_test,y_pred))
print(f1_score(y_test,y_pred))
```

```
prediction=model.predict(test_new)
res = pd.DataFrame(prediction)
res.index = test_new.index # its important for comparison
res.columns = ["prediction"]

# To download the csv file locally
from google.colab import files
res.to_csv('prediction_results_HP.csv')         
files.download('prediction_results_HP.csv')
```
